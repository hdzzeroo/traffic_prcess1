# 2. 各区間の2kmトラカン基数を列挙

```python
# TARGET_ROAD = 'kannetsu'
TARGET_ROAD = 'touhoku'

PERIOD = '20210401-20220331'

RAW_CSV = f'./traffic/raw_{TARGET_ROAD}_{PERIOD}_2KP.csv'
PROCESSED_CSV = f'./traffic/{TARGET_ROAD}_{PERIOD}_2KP.csv'

NUM_CSV = f'../../Input_processed_data/road_master/num_traffic_counter_2KP.csv'
```

## マスタデータ

```python
df_master = pd.read_csv(NUM_CSV)
df_master.head()
```

```python
if TARGET_ROAD == 'kannetsu':
    df_master = df_master.loc[df_master.道路 == '関越自動車道'].reset_index(drop=True)
elif TARGET_ROAD == 'touhoku':
    df_master = df_master.loc[df_master.道路 == '東北自動車道'].reset_index(drop=True)
```

### 区間名称の分割

```python
def split_segment_name(df):
    # 全角スペースを除去 → "〜"で分割
    res = df.assign(segment_name=(df.区間名称.str.replace('\u3000', '').str.replace('\u3000', '').str.split('〜')))
    res = res.assign(
        start_name=res.segment_name.apply(lambda l: l[0]),
        end_name=res.segment_name.apply(lambda l: l[1])
    )
    return res
```

```python
df_master = split_segment_name(df_master)
df_master.head()
```

### 上り区間・下り区間に応じて発着IC名を入れ替える
= 上り区間の発着IC名を入れ替える

```python
def swap_start_end(df):
    up_start = df.loc[df.方向 == '上り', 'start_name']
    up_end = df.loc[df.方向 == '上り', 'end_name']

    df.loc[df.方向 == '上り', 'start_name'] = up_end
    df.loc[df.方向 == '上り', 'end_name'] = up_start
    
    return df
```

```python
df_master = swap_start_end(df_master)
df_master.head()
```

### IC名称をICマスタデータに統一

```python
IC_CSV = f'../../Input_processed_data/road_master/ic_preprocessed.csv'
IC_SUBNET_CSV = f'../../Input_processed_data/road_master/icnet_sub.csv'
```

```python
df_ic = pd.read_csv(IC_CSV)
df_icnet = pd.read_csv(IC_SUBNET_CSV, dtype={'start_code': str, 'end_code': str, 'road_code': str})

# name => code
name2code = dict(zip(df_ic['ic_name'], df_ic['ic_code']))
name_traffic2code = dict(zip(df_ic['ic_name_traffic'], df_ic['ic_code']))
name2code.update(name_traffic2code)

# code => name
code2name = dict(zip(df_ic['ic_code'], df_ic['ic_name']))
```

```python
def convert_ic_name(df):
    # IC名をICコードに直す（トラカン固有のIC名を含む）
    res = df.assign(
        direction=df.方向.map({'上り': 'up', '下り': 'down'}),
        start_code=df.start_name.map(name2code),
        end_code=df.end_name.map(name2code))
    
    # ICコードからIC名称に変換
    res = res.assign(
        start_name=res.start_code.map(code2name),
        end_name=res.end_code.map(code2name))
    
    return res
```

```python
df_master = convert_ic_name(df_master)
df_master.head()
```

```python
df_master = df_master.loc[:, ['区間名称', '方向', 'start_name', 'start_code', 'end_name', 'end_code', '基数']]
df_master.head()
```
