# ICネットワークデータの処理

## やっていること
- 館山道・関越道のICネットワーク表と東北道のICネットワーク表を結合

## Input
- 館山道・関越道のICネットワーク表: `Input_processed_data/road_master/tateyama_kannetsu_icnet.csv`
- 東北道のICネットワーク表: `Input_processed_data/road_master/touhoku_icnet.csv`

## Output
- 予測用に限定されたICネットワーク表: `Input_processed_data/road_master/icnet_sub.csv`

```python
import pandas as pd
```

```python
# data directory
PROCESSED_DATA_DIR = '../../Input_processed_data'
```

```python
df_kannetsu = pd.read_csv(f'{PROCESSED_DATA_DIR}/road_master/tateyama_kannetsu_icnet.csv')
df_touhoku = pd.read_csv(f'{PROCESSED_DATA_DIR}/road_master/touhoku_icnet.csv')
```

```python
drop_columns = ['type', 'kukan_name', 'kukan_no']

df_kannetsu.drop(drop_columns, axis=1, inplace=True)
```

```python
(df_kannetsu.columns == df_touhoku.columns).all()
```

```python
df_subnet = pd.concat([df_kannetsu, df_touhoku]).reset_index(drop=True)
df_subnet
```
